```python
class recipe(Fact):
    pass
```

```python
recipes=[]
for i in range(10000):
    r=recipe(title=df['title'][i],nutritions=ast.literal_eval(df['nutritions'][i]))
    if (r['nutritions']['calories']!=0) and (r['nutritions']['cholesterolContent']!=0) and (r['nutritions']['fatContent']!=0) and (r['nutritions']['proteinContent']!=0) :
        recipes.append(r)
```

```python
calories_limit=700                   ## 2000
carbohydrate_limit=400               ## 1200
cholesterol_limit=100                ## 300
Fat_limit=70                         ## 100
saturatedFat_limit=15                 ## 15
sodium_limit=500                     ## 1500
sugar_limit=10                       ## 30
protein_limit=70                     ## 200
```

```python
class Systolic(Fact):
    pass
class Diastolic(Fact):
    pass
class Disease(Fact):
    pass
class Pressure(KnowledgeEngine):
    @DefFacts()
    def initialfacts(self):
        yield Fact(calories_limit=700)
        yield Fact(carbohydrate_limit=400)
        yield Fact(cholesterol_limit=100)
        yield Fact(Fat_limit=70 )
        yield Fact(saturatedFat_limit=15)
        yield Fact(sodium_limit=500)
    @Rule(NOT(Fact(Systolic = W())))
    def enter(self):
        s1=Systolic(measure='first',m=input("enter the  test result for systolic pressure for the first measure "))
        d1=Diastolic(measure='first',m=input("enter the test result for diastolic pressure for the first measure "))
        self.declare(s1)
        self.declare(d1)
        
        s2=Systolic(measure='second',m=input("enter the  test result for systolic pressure for the second measure "))
        d2=Diastolic(measure='second',m=input("enter the test result for diastolic pressure for the second measure "))
        self.declare(s2)
        self.declare(d2)
        
        s3=Systolic(measure='third',m=input("enter the  test result for systolic pressure for the third measure "))
        d3=Diastolic(measure='third',m=input("enter the test result for diastolic pressure for the third measure "))
        self.declare(s3)
        self.declare(d3)
        
        self.declare(Systolic(measure='result',m=(int(s1['m'])+int(s2['m'])+int(s3['m']))/3))
        self.declare(Diastolic(measure='result',m=(int(d1['m'])+int(d2['m'])+int(d3['m']))/3))
        
        
    @Rule(AND(Systolic(measure=L('result'),m=P(lambda x: int(x) <= 120 ))
              ,Diastolic(measure=L('result'),m=P(lambda x: int(x) <= 80 ))))
    def result(self):
        self.declare(Disease(name="NONE"))
        print('you can eat what you want')
        
        
    @Rule(AND(Systolic(measure=L('result'),m=P(lambda x: int(x) < 130 ) & P(lambda x: int(x) > 120 ))
              ,Diastolic(measure=L('result'),m=P(lambda x: int(x) < 80 ))))
    def result2(self):
        self.declare(Disease(name="Hybertension"))
        print('elevated')
    
    
    @Rule(OR(Systolic(measure=L('result'),m=P(lambda x: int(x) < 140 ) & P(lambda x: int(x) >= 130 ))
             ,Diastolic(measure=L('result'),m=P(lambda x: int(x) < 90 ) & P(lambda x: int(x) > 80 ))),NOT(Disease(name='a'<<W())))
    def result3(self):
        self.declare(Disease(name="Hybertension S1"))
        self.modify(self.facts[1], calories_limit=0.90*calories_limit)
        self.modify(self.facts[2], carbohydrate_limit=0.90*carbohydrate_limit)
        self.modify(self.facts[3], cholesterol_limit=0.90*cholesterol_limit)
        self.modify(self.facts[4], Fat_limit=0.90*Fat_limit)
        self.modify(self.facts[5], saturatedFat_limit=0.90*saturatedFat_limit)
        self.modify(self.facts[6], sodium_limit=0.90*sodium_limit)
    
    @Rule(OR(Systolic(measure=L('result'),m=P(lambda x: int(x) >= 140 ))
         ,Diastolic(measure=L('result'),m=P(lambda x: int(x) >= 90 ))),salience=10)
    def result4(self):
        self.declare(Disease(name="Hybertension S2"))
        self.modify(self.facts[1], calories_limit=0.80*calories_limit)
        self.modify(self.facts[2], carbohydrate_limit=0.80*carbohydrate_limit)
        self.modify(self.facts[3], cholesterol_limit=0.80*cholesterol_limit)
        self.modify(self.facts[4], Fat_limit=0.80*Fat_limit)
        self.modify(self.facts[5], saturatedFat_limit=0.80*saturatedFat_limit)
        self.modify(self.facts[6], sodium_limit=0.80*sodium_limit)
        
    @Rule(Disease(name=P(lambda x: x=='Hybertension' ) | P(lambda x: x=='Hybertension S1' ) | P(lambda x: x=='Hybertension S2' ))
            , Fact(calories_limit='a'<<W())
            , Fact(carbohydrate_limit='b'<<W())
            , Fact(cholesterol_limit='c'<<W())
            , Fact(Fat_limit='d'<<W())
            , Fact(saturatedFat_limit='e'<<W())
        , Fact(sodium_limit='f'<<W()) )
    def action1(self,a,b,c,d,e,f):
        print(self.facts)
        ans=input("do you have Diabetes ? (yes/no)")
        if ans=='yes':
            self.declare(Disease(name="Diabetes"))
            b-=0.10*b
        ans=input("do you have kidney disease ? (yes/no)")
        if ans=='yes':
            self.declare(Disease(name="kidney disease"))
            c-=0.10*c    
            d-=0.10*d 
            e-=0.10*e
        
        rs=[]
        for i in range(len(recipes)):
            if recipes[i]['nutritions']['calories']<a and recipes[i]['nutritions']['carbohydrateContent']<b:
                if recipes[i]['nutritions']['cholesterolContent']<c and recipes[i]['nutritions']['fatContent']<d:
                    if recipes[i]['nutritions']['saturatedFatContent']<e and recipes[i]['nutritions']['sodiumContent']<f:
                        rs.append(recipes[i])
        rs=sorted(rs,key=lambda x: (x['nutritions']['sodiumContent'],x['nutritions']['saturatedFatContent']
                                    ,x['nutritions']['fatContent'],x['nutritions']['cholesterolContent']
                                    ,x['nutritions']['carbohydrateContent'],x['nutritions']['calories']), reverse=False)
        self.declare(rs[0])
        self.declare(rs[1])
        self.declare(rs[2])
        for i in range(7):
             self.declare(rs[random.randint(3,len(rs))])
            
    
```

```python

i=Pressure()

i.reset()

i.run()
```
